Random Walks - using pythong to generate data from a random walk then using matplotlib to visualize  the results

```python
from random import choice

# 1. Creating a random walking class
class Randomwalk:
    """A Class to generate random walks data."""

    def __init__(self, num_points=5000):
        """Initalize attributes of a walk."""
        self.num_points = num_points

        # All walks start at (0,0).
        self.x_values = [0]
        self.y_values = [0]

# 2. Choosing directions 
    def fill_walk(self):
        """Calculate all the points in the walk."""
        #Keep taking steps until the walk reaches the desired length/
        while len(self.y_values) <self.num_points: 

            #Decide which directions to go, and how far to go.
            x_direction =choice([1,-1])
            x_distance= choice([0,1,2,3,4])
            x_step = x_direction*x_distance

            y_direction = choice ([1,-1])
            y_distance = choice ([0,1,2,3,4])
            y_step = y_direction*y_distance

            #Reject Moves that go nowhere. 
            if x_step == 0 and y_step ==0: 
                continue

            #Calculate the new position. 
            x= self.x_values[-1]+ x_step
            y=self.y_values[-1]+ y_step

            self.x_values.append(x)
            self.y_values.append(y)

```

```python
import matplotlib.pyplot as plt


rw = Randomwalk()
rw.fill_walk()

# Plot the points in the walk. 
plt.style.use('classic')
fig,ax = plt.subplots()
ax.scatter(rw.x_values,rw.y_values, s=15)
ax.set_aspect('equal')

plt.show()
```

```python
from random import choice

# 1. Creating a random walking class
class Randomwalk:
    """A Class to generate random walks data."""

    def __init__(self, num_points=5000):
        """Initalize attributes of a walk."""
        self.num_points = num_points

        # All walks start at (0,0).
        self.x_values = [0]
        self.y_values = [0]

# 2. Choosing directions 
    def fill_walk(self):
        """Calculate all the points in the walk."""
        #Keep taking steps until the walk reaches the desired length/
        while len(self.y_values) <self.num_points: 

            #Decide which directions to go, and how far to go.
            x_direction =choice([1,-1])
            x_distance= choice([0,1,2,3,4])
            x_step = x_direction*x_distance

            y_direction = choice ([1,-1])
            y_distance = choice ([0,1,2,3,4])
            y_step = y_direction*y_distance

            #Reject Moves that go nowhere. 
            if x_step == 0 and y_step ==0: 
                continue

            #Calculate the new position. 
            x= self.x_values[-1]+ x_step
            y=self.y_values[-1]+ y_step

            self.x_values.append(x)
            self.y_values.append(y)
```

```python
import plotly.express as px

rw = Randomwalk() 
rw.fill_walk()

#visualize the random walk
title = "Random walk" 
labels = {'x':'steps','y': 'directions'}
fig = px.scatter(x=rw.x_values, y=rw.y_values,title=title,labels=labels)
fig.show() 
```
