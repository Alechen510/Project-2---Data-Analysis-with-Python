ticket_label method allow your to overide the the default tick labels
Using colormap

```python
import matplotlib.pyplot as plt

x_values = range(1, 1001)
y_values =[x**2 for x in x_values]

plt.style.use('fivethirtyeight')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s=10)

#set chart tittle and label axis 
ax.set_title("Square Numbers Chart", fontsize =14)
ax.set_xlabel("Values", fontsize = 14)
ax.set_ylabel("Square Numbers", fontsize = 14)
ax.tick_params(labelsize = 14)


# set the range for each axis
ax.axis  ([0,1100, 0, 1_100_000])
# Set the the ticklabel_format
ax.ticklabel_format(style ='plain')

plt.show()
```

Customizing Tick Labels 
Definining customer colors 
Using Color map 

```python
import matplotlib.pyplot as plt

# x and y values and calculation
x_values =  range(1,1001)
y_values = [x**2 for x in x_values]

#choose the style of the scatter plot 
plt.style.use('fivethirtyeight')
fig,ax = plt.subplots()
ax.scatter(x_values,y_values, c=y_values, cmap= plt.cm.Reds, s=10)

# Set chart labels and lable axis 
ax.set_title("Square Numbers Chart", fontsize = 14)
ax.set_xlabel("Values", fontsize = 14)
ax.set_ylabel("Square Numbers", fontsize = 14)
ax.tick_params(labelsize = 14)

#Set the range for each axis
ax.axis = ([0,1100, 0, 1_100_000])
ax.ticklabel_format(style='plain')

plt.show()
```

Saving Your Plots Automatically using plt.savefig() instead of plt.show()

```python
import matplotlib.pyplot as plt

# x and y values and calculation
x_values =  range(1,1001)
y_values = [x**2 for x in x_values]

#choose the style of the scatter plot 
plt.style.use('fivethirtyeight')
fig,ax = plt.subplots()
ax.scatter(x_values,y_values, c=y_values, cmap= plt.cm.Reds, s=10)

# Set chart labels and lable axis 
ax.set_title("Square Numbers Chart", fontsize = 14)
ax.set_xlabel("Values", fontsize = 14)
ax.set_ylabel("Square Numbers", fontsize = 14)
ax.tick_params(labelsize = 14)

#Set the range for each axis
ax.axis = ([0,1100, 0, 1_100_000])
ax.ticklabel_format(style='plain')

plt.savefig('Squares_plot.png',bbox_inches = 'tight')
```

15-1. Cubes: A number raised to the third power is a cube. Plot the first five
cubic numbers, and then plot the first 5,000 cubic numbers.

```python
import matplotlib.pyplot as plt

x_values = [1,2,3,4,5]
y_values = [x**3 for x in x_values]

plt.style.use('fivethirtyeight')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, color = 'Green', s=10)

#setting title and x y labels 
ax.set_title("Cubic Numbers Chart", fontsize = 14)
ax.set_xlabel("Values", fontsize = 14)
ax.set_ylabel("Cubic Numbers", fontsize = 14)
ax.tick_params(labelsize = 10)

#set the range for each axis 
#ax.axis([0,5, 1,200])
#set the ticklabel_format
#ax.ticklabel_format(style='plain')

plt.show()
```

```python
import matplotlib.pyplot as plt

x_values = range(1, 5001)
y_values = [x**3 for x in x_values]

plt.style.use('fivethirtyeight')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, c=y_values, cmap =plt.cm.Greens, s=10)

#setting title and x y labels 
ax.set_title("Cubic Numbers Chart", fontsize = 14)
ax.set_xlabel("Values", fontsize = 14)
ax.set_ylabel("Cubic Numbers", fontsize = 14)
ax.tick_params(labelsize = 10)

#set the range for each axis 
ax.axis([0,5100, 1,125_000_000_000])
#set the ticklabel_format
ax.ticklabel_format(style='plain')

plt.show()
```
